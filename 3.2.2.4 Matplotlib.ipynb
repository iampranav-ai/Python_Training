```python
df1 = pd.DataFrame({
    "Sid": np.arange(1, 9),
    "Maths": np.random.randint(50, 100, 8),
    "Sci": np.random.randint(50, 100, 8)
})
```

```python
plt.bar(df1['Sid'], df1['Maths'], label="MATHS", alpha=0.5, width=0.5)
plt.bar(df1['Sid'], df1['Sci'], label="SCIENCE", alpha=0.5, width=0.5,
        align='edge')
plt.legend()
```

```python
plt.bar(df1['Sid'], df1['Maths'], label="MATHS", alpha=0.5, width=0.5)
plt.bar(df1['Sid'], df1['Sci'], label="SCIENCE", alpha=0.5, width=0.5,
        align='edge', bottom=df1['Maths'])
plt.legend()
```
