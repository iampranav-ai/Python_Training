```python
ticks = {'fontsize': 12,'color':'red', 'fontfamily':'monospace', 'fontstyle': 'italic'}
```

```python
plt.plot(df2['Sci'], '1-.',color='red', mec='k', ms=15, mfc='k', lw=4)
plt.plot(df2['Maths'], 'x:', color="green", mec='k', ms=15, mfc='k')
plt.xticks(color='purple', fontsize=15, fontfamily='monospace',fontstyle='italic')
plt.yticks(color='purple', fontsize=15, fontfamily='monospace',fontstyle='italic')
plt.title("MARKS", fontdict=ticks)
plt.show()
```

```python
plt.plot(df2['Sci'], '1-.',color='red', mec='k', ms=15, mfc='k', lw=4)
plt.plot(df2['Maths'], 'x:', color="green", mec='k', ms=15, mfc='k')
plt.tick_params(labelcolor='purple', labelsize=12, size=10, width=5)
plt.title("MARKS", fontdict=ticks)
plt.show()
```

```python
plt.figure(facecolor='black')
plt.plot(df2['Sci'], '1-.', color='red', mec='white', ms=15, mfc='k', lw=4)
plt.plot(df2['Maths'], 'x:', color='green', mec='white', ms=15, mfc='k', lw=4)
plt.tick_params(labelcolor='white',labelsize=12, size=10,width=5,color='white')
plt.title("MARKS", fontdict=ticks)
ax=plt.gca()
ax.set_facecolor("black")
ax.spines['left'].set_color("white")
ax.spines['bottom'].set_color("white")
plt.show()
```

```python
plt.figure(facecolor='black')
plt.plot(df2['Sci'], '1-.', color='red', mec='white', ms=15, mfc='k',
        lw=4, label='SCI')
plt.plot(df2['Maths'], 'x:', color='green', mec='white', ms=15, mfc='k',
        lw=4, label="MATHS")
plt.tick_params(labelcolor='white', labelsize=12, size=10,
               width=5, color='white')
plt.title("MARKS", fontdict=ticks)
ax = plt.gca()
ax.set_facecolor("black")
spines = ['left', 'right', 'top', 'bottom']
for sp in spines:
    ax.spines[sp].set_color("white")
    ax.spines[sp].set_ls("--")
plt.legend()
plt.grid()
plt.show()
```
