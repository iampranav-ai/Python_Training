```python
def filter_(x):
    if type(x) == str:
        if x[0].lower() == 'a':
            return True
    return False
```

```python
taxi['pickup_zone'].apply(filter_).sum()
```

```python
taxi[taxi['pickup_zone'].apply(filter_)]['pickup_zone'].unique()
```

```python
taxi[taxi['pickup_zone'].apply(filter_)]['pickup_zone'].nunique()
```
