```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tip = sns.load_dataset("tips")
titanic = sns.load_dataset("titanic")
taxi = sns.load_dataset("taxis")
```

```python
taxi['passengers'].value_counts()
```

```python
plt.rcParams['figure.figsize'] = 10, 5
plt.scatter(taxi['distance'],taxi['fare'],marker='*',c=taxi['color'])
```

```python
i1=taxi['passengers'].value_counts().index
v1=taxi['passengers'].value_counts()
c = ['red','green','blue','#543bcd','yellow']
plt.bar(i1,v1,color=c)
```

```python
i1=tip['day'].value_counts().index
v1=tip['day'].value_counts()
c = ['red','green','blue','yellow']
plt.bar(i1,v1,color=c,hatch='|')
```

```python
i1=tip['day'].value_counts().index
v1=tip['day'].value_counts()
c = ['red','green','blue','yellow']
plt.bar(i1,v1,color=c,hatch='x')
```

```python
i1=tip['day'].value_counts().index
v1=tip['day'].value_counts()
c = ['red','green','blue','yellow']
plt.bar(i1,v1,color=c,hatch='|',width=0.5,ec='purple',ls='--')
```

```python
i1=tip['day'].value_counts().index
v1=tip['day'].value_counts()
i2=tip['time'].value_counts().index
v2=tip['time'].value_counts()
c = ['red','green','blue','yellow']
plt.bar(i1,v1,color=c,hatch='|',width=0.5,ec='purple',ls='--')
plt.bar(i2,v2,color=c,hatch='|',width=0.5,ec='purple',ls='--')
```
